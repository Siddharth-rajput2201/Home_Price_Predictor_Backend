```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize']=(20,10)
```

```python
df1 = pd.read_csv('MagicBricks.csv')
df1
```

```python
df2 = df1.drop(['Per_Sqft'],axis ='columns')

df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3['Bathroom'].unique()
```

```python
df3[df3.Bathroom == 7]
```

```python
df3
```

```python
df3["Area"] = df3["Area"].astype(int)
df3
```

```python
df4 = df3.copy()
df4['price_per_sqft'] = df4['Price']/df4['Area']
df4.head()
```

```python
len(df4.Area.unique())
```

```python
df4.Locality = df4.Locality.apply(lambda x: x.strip())
```

```python
location_stats = df4.groupby('Locality')['Locality'].agg('count').sort_values(ascending = False)
print(location_stats)
print('\n')
```

```python
df4["value"] = df4["Locality"].str.len() > 50
```

```python
df5 = df4.copy()
df5
```

```python
df5 = df4.drop(df5.index[df5['value']])
df5
```

```python
df6 = df5.drop(['value'], axis = 'columns')
df6
```

```python
location_stats = df6.groupby('Locality')['Locality'].agg('count').sort_values(ascending = False)
location_stats
```

```python
len(location_stats[location_stats<=5])
```

```python
location_stats_less_than_3 = location_stats[location_stats <= 3]
location_stats_less_than_3
```

```python
df6.Locality = df6.Locality.apply(lambda x: 'other' if x in location_stats_less_than_3 else x)
len(df6.Locality.unique())
```

```python
df6[df6.Area/df5.BHK<300].head()
```

```python
df6.head()
```

```python
df7 = df6.drop(['price_per_sqft','Status','Transaction','Type','Furnishing'],axis = 'columns')
df7.head()
```

```python
dummies = pd.get_dummies(df7.Locality)
```

```python
dummies.head()
```

```python
df8 = pd.concat([df7,dummies.drop('other',axis='columns')],axis = 'columns' )
df8.head()
```

```python
df9 = df8.drop('Locality',axis = 'columns')
df9.head()
```

```python
df9.shape
```

```python
x = df9.drop('Price',axis='columns')
x.head()
```

```python
y = df9.Price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.1, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
loc_index = np.where(x.columns=='Rohini Sector 24')[0][0]
loc_index
```

```python
def predict_price(location,area,bathroom,bhk,parking):
    x = df9.drop('Price',axis='columns')
    loc_index = np.where(x.columns==location)[0][0]

    x = np.zeros(len(x.columns))
    x[0] = area
    x[1] = bathroom
    x[2] = bhk
    x[3] = parking
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```

```python
predict_price('Rohini Sector 24',1000,3,3,1)
```
